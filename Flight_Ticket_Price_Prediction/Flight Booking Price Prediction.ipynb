```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ['airline', 'flight', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']
```

```python
encoders = {}
for col in categorical_cols:
    encoders[col] = LabelEncoder()
```

```python
for col in categorical_cols:
    data[col] = encoders[col].fit_transform(data[col])
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(data.corr(), annot=True, cmap='RdBu_r', vmin=-1, vmax=1,annot_kws={"fontsize":8})
plt.show()
```

```python
X = data.drop('price',axis=1)
y = data.price
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in data.columns:
    if ((data[col].dtype !='object') & (col != 'price') ):
        col_list.append(col)
        
x = data[col_list] 
vif_data = pd.DataFrame()
vif_data["feature"]=x.columns
vif_data["VIF"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
print(vif_data)
```

vif value of flight is very hight

```python
X.drop('flight',axis=1,inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=15)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = DecisionTreeRegressor().fit(X_train,y_train)
y_pred = model1.predict(X_test)
```

```python
from sklearn import metrics
print(f"Mean Abslote Error : {metrics.mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared Error : {metrics.mean_squared_error(y_test,y_pred)}")
print(f"Root Mean Squared Error : {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}")
print(f"R^2 : {metrics.r2_score(y_test,y_pred)}")
```

```python
model2 = RandomForestRegressor(n_estimators= 500).fit(X_train,y_train)
y_pred2 = model2.predict(X_test)
```

```python
print(f"Mean Abslote Error : {metrics.mean_absolute_error(y_test,y_pred2)}")
print(f"Mean Squared Error : {metrics.mean_squared_error(y_test,y_pred2)}")
print(f"Root Mean Squared Error : {np.sqrt(metrics.mean_squared_error(y_test,y_pred2))}")
print(f"R^2 : {metrics.r2_score(y_test,y_pred2)}")
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.distplot(y_test, label="Actual", ax=ax)
sns.distplot(y_pred2, label="Predicted", ax=ax, color="red")
plt.legend()
plt.show()
```

```python
sns.regplot(y_pred2,y_test,line_kws={'color': 'blue'},color='lightblue')
plt.show()
```
